```python
# explotarery data Analysis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['ssc_percentage'], bins=20)
plt.title("ssc_percentage  Data Destribution")
plt.xlabel("percentage")
plt.ylabel("data count")
plt.show()
```

```python
plt.hist(df['hsc_percentage'])
plt.title("hsc percentage data distribution")
plt.xlabel("percentage")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(df['gender'])
plt.title("gender data distribution")
plt.xlabel("percentage")
plt.ylabel("count")
plt.show()
```

```python
sns.scatterplot(x='ssc_percentage', y='hsc_percentage', data =df,hue='status')
plt.title("Difference bw ssc and hsc student")
plt.xlabel("SSC Percentage")
plt.ylabel("HSC percentage")
plt.show()
```
